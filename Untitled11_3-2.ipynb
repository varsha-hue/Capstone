```python
CSV_PATH = "/content/crc_dataset.csv"

HIDDEN = 256
DROPOUT = 0.2
K = 10
LR = 0.002
WEIGHT_DECAY = 1e-3
MAX_EPOCHS = 800
PATIENCE = 80

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
df = pd.read_csv(CSV_PATH)

# ✅ NORMALIZE COLUMNS IMMEDIATELY AFTER LOADING
df.columns = (
    df.columns
    .str.strip()
    .str.replace("-", "_")
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)

feature_cols = [c for c in df.columns if c not in ("Participant_ID", "CRC_Risk")]
target_col = "CRC_Risk"
```

```python
num_cols = df[feature_cols].select_dtypes(include=["int64", "float64"]).columns.tolist()
scaler = StandardScaler()
X_num = scaler.fit_transform(df[num_cols]) if len(num_cols) > 0 else np.zeros((len(df),0))
cat_cols = [c for c in feature_cols if c not in num_cols]
if len(cat_cols) > 0:
    try:
        enc = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    except TypeError:
        enc = OneHotEncoder(sparse=False, handle_unknown="ignore")
    X_cat = enc.fit_transform(df[cat_cols].astype(str))
else:
    X_cat = np.zeros((len(df),0))
X = np.hstack([X_num, X_cat]).astype(np.float32)
y = df[target_col].values.astype(int)
```

```python
n_qubits = X_q.shape[1]
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_feature_map(x):
    for i in range(n_qubits):
        qml.RY(x[i], wires=i)
    for i in range(n_qubits - 1):
        qml.CNOT(wires=[i, i + 1])
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

X_quantum = np.array([quantum_feature_map(x) for x in X_q])
print("Quantum feature shape:", X_quantum.shape)
```

```python
nbrs = NearestNeighbors(n_neighbors=K+1, metric="cosine").fit(X)
_, indices = nbrs.kneighbors(X)

edges = set()
for i in range(len(X)):
    for j in indices[i, 1:]:
        if i in indices[j, 1:]:
            edges.add((min(i, j), max(i, j)))

edge_index = torch.tensor(list(edges), dtype=torch.long).t().contiguous()
print(f"Graph: {len(X)} nodes, {edge_index.size(1)} undirected edges")
```

```python
alpha = 0.2 # try 0.1 → 0.3
X_quantum_scaled = alpha * X_quantum
X_final = np.hstack([X, X_quantum_scaled])
x = torch.tensor(X_final, dtype=torch.float)
y_tensor = torch.tensor(y, dtype=torch.long)
data = Data(x=x, edge_index=edge_index, y=y_tensor)
idx = np.arange(len(X))
train_idx, test_idx = train_test_split(idx, test_size=0.2, stratify=y, random_state=42)
train_idx, val_idx = train_test_split(train_idx, test_size=0.125, stratify=y[train_idx], random_state=42)
train_mask = torch.zeros(len(X), dtype=torch.bool); train_mask[train_idx] = True
val_mask   = torch.zeros(len(X), dtype=torch.bool); val_mask[val_idx] = True
test_mask  = torch.zeros(len(X), dtype=torch.bool); test_mask[test_idx] = True
data.train_mask, data.val_mask, data.test_mask = train_mask, val_mask, test_mask
```

```python
class GraphSAGEModel(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, dropout=DROPOUT):
        super().__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, hidden_channels)
        self.conv3 = SAGEConv(hidden_channels, hidden_channels)
        self.lin   = nn.Linear(hidden_channels, num_classes)
        self.dropout = dropout
    def forward(self, x, edge_index):
        x1 = F.relu(self.conv1(x, edge_index))
        x1 = F.dropout(x1, p=self.dropout, training=self.training)
        x2 = F.relu(self.conv2(x1, edge_index))
        x2 = F.dropout(x2, p=self.dropout, training=self.training)
        x3 = self.conv3(x2, edge_index) + x1
        return self.lin(x3)
```

```python
class GATModel(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, dropout=DROPOUT):
        super().__init__()
        self.gat1 = GATConv(in_channels, hidden_channels, heads=4, concat=True, dropout=dropout)
        self.gat2 = GATConv(hidden_channels*4, hidden_channels, heads=1, concat=False, dropout=dropout)
        self.lin  = nn.Linear(hidden_channels, num_classes)
        self.dropout = dropout
    def forward(self, x, edge_index):
        x = F.elu(self.gat1(x, edge_index))
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = F.elu(self.gat2(x, edge_index))
        return self.lin(x)
```

```python
def train_model(model, data):
    model = model.to(DEVICE)
    data = data.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
    criterion = nn.CrossEntropyLoss(label_smoothing=0.03)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode="max", factor=0.5, patience=20
    )

    best_val, best_state = 0, None
    patience_counter = 0
    for epoch in range(1, MAX_EPOCHS + 1):
        model.train()
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = criterion(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
        model.eval()
        with torch.no_grad():
            pred = out[data.val_mask].argmax(dim=1)
            acc = accuracy_score(data.y[data.val_mask].cpu(), pred.cpu())
        scheduler.step(acc)
        if acc > best_val:
            best_val = acc
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            patience_counter = 0
        else:
            patience_counter += 1
        if epoch % 25 == 0:
            print(f"Epoch {epoch:03d}, Loss {loss:.4f}, Val Acc {acc:.4f}")
        if patience_counter >= PATIENCE:
            print(f"Early stopping at epoch {epoch}, best val acc={best_val:.4f}")
            break
    if best_state is not None:
        model.load_state_dict(best_state)
    return model
```

```python
print("\n Training GraphSAGE ")
sage_model = GraphSAGEModel(data.num_features, HIDDEN, len(np.unique(y)))
sage_model = train_model(sage_model, data)
sage_model.eval()
out = sage_model(data.x.to(DEVICE), data.edge_index.to(DEVICE))
pred_sage = out[data.test_mask].argmax(dim=1).cpu()
sage_acc = accuracy_score(data.y[data.test_mask].cpu(), pred_sage)
print("\nGraphSAGE Test Accuracy:", sage_acc)
```

```python
print("\n Training GAT ")
gat_model = GATModel(data.num_features, HIDDEN, len(np.unique(y)))
gat_model = train_model(gat_model, data)
gat_model.eval()
out = gat_model(data.x.to(DEVICE), data.edge_index.to(DEVICE))
pred_gat = out[data.test_mask].argmax(dim=1).cpu()
gat_acc = accuracy_score(data.y[data.test_mask].cpu(), pred_gat)
print("\nGAT Test Accuracy:", gat_acc)
```

```python
print("\n Ensemble (GraphSAGE + GAT) ")
out_sage = sage_model(data.x.to(DEVICE), data.edge_index.to(DEVICE))
out_gat  = gat_model(data.x.to(DEVICE), data.edge_index.to(DEVICE))
out_ens  = out_ens = 0.3 * out_gat + 0.7 * out_sage
pred_ens = out_ens[data.test_mask].argmax(dim=1).cpu()
ens_acc = accuracy_score(data.y[data.test_mask].cpu(), pred_ens)
print("\nEnsemble Test Accuracy:", ens_acc)
print("\nClassification Report (Ensemble on Test):\n",
      classification_report(data.y[data.test_mask].cpu(), pred_ens, digits=4))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import NearestNeighbors
import torch
from torch_geometric.data import Data
df = pd.read_csv('crc_dataset.csv')
cat_cols = ['Gender','Lifestyle','Ethnicity','Family_History_CRC','Pre-existing Conditions']
for c in cat_cols:
    df[c] = LabelEncoder().fit_transform(df[c].astype(str))
feature_cols = [c for c in df.columns if c not in ['Participant_ID','CRC_Risk']]
X = df[feature_cols].values
y = df['CRC_Risk'].values
participant_ids = df['Participant_ID'].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
k = 10
nbrs = NearestNeighbors(n_neighbors=k+1, metric='cosine').fit(X_scaled)
_, idx = nbrs.kneighbors(X_scaled)
edges = set()
for i in range(len(X_scaled)):
    for j in idx[i,1:]:
        if i in idx[j,1:]:
            a,b = (i,j) if i<j else (j,i)
            edges.add((a,b))
edge_index = torch.tensor(list(edges), dtype=torch.long).t().contiguous()
x = torch.tensor(X_scaled, dtype=torch.float32)
y_t = torch.tensor(y, dtype=torch.long)
pyg_data = Data(x=x, edge_index=edge_index, y=y_t)
pyg_data.participant_ids = torch.tensor(participant_ids)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from torch_geometric.utils import to_networkx
import random
G = to_networkx(pyg_data, to_undirected=True)
sample_size = 30
sample_nodes = random.sample(list(G.nodes()), min(sample_size, G.number_of_nodes()))
sub_G = G.subgraph(sample_nodes)
crc_risks = pyg_data.y.cpu().numpy()
ids = pyg_data.participant_ids.cpu().numpy()
node_color = [crc_risks[n] for n in sub_G.nodes()]
labels = {n: str(ids[n]) for n in sub_G.nodes()}
plt.figure(figsize=(10,8))
nx.draw(sub_G, node_color=node_color, labels=labels, with_labels=True,
        node_size=400, font_size=8, cmap=plt.cm.Set1)
plt.title(f"GNN Visualization (Random {len(sub_G.nodes())} Nodes): Color=CRC Risk, Label=Participant ID")
plt.show()
```
